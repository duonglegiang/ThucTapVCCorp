```python
from PIL import *
import pytesseract
import cv2
from cv2 import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read img
im = cv2.imread("/home/giang/Desktop/Test/TestML/xe10.jpg", 1)
plt.imshow(cv2.cvtColor(im,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# chuyen sang anh xam
im_gray = cv2.imread("/home/giang/Desktop/Test/TestML/xe10.jpg", 0)
plt.imshow(cv2.cvtColor(im_gray,cv2.COLOR_BGR2RGB))
plt.title('anh_xam')
plt.show()
```

```python
img_gray2 = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(img_gray2,cv2.COLOR_BGR2RGB))
plt.title('anh xam')
plt.show()
```

```python
# resize anh
img_resize = cv2.resize(im, (200, 200))
plt.imshow(cv2.cvtColor(img_resize,cv2.COLOR_BGR2RGB))
plt.title('img (200 * 200)')
plt.show()
```

```python
# crop anh
img_crop = im[240:400, 250:460, :]
plt.imshow(cv2.cvtColor(img_crop,cv2.COLOR_BGR2RGB))
plt.title('img crop')
plt.show()
```
